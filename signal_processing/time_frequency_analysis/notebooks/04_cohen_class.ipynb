## Cohen's Class: A Unified Framework for Time-Frequency Distributions
## Cohenのクラス：時間周波数分布の統一的フレームワーク

---
> **Reference:** Cohen, *Time-Frequency Analysis*, Ch. 6–8; Cohen (1966).

## 4.1 The Unification Idea
## 4.1 統一の発想

**English.**  
Cohen (1966) showed that **any** bilinear time-frequency distribution satisfying time and frequency covariance can be written as:

$$\boxed{C_x(t, f) = \iint A_x(\eta, \tau)\, \Phi(\eta, \tau)\, e^{j2\pi(\eta t - f\tau)}\, d\eta\, d\tau}$$

where:
- $A_x(\eta, \tau)$ is the **symmetric ambiguity function** of $x$:
  $$A_x(\eta, \tau) = \int x\!\left(t + \frac{\tau}{2}\right) x^*\!\left(t - \frac{\tau}{2}\right) e^{-j2\pi\eta t}\, dt$$
- $\Phi(\eta, \tau)$ is the **kernel function** (filter in the ambiguity domain)

Every choice of $\Phi$ yields a different member of **Cohen's class**.  
The kernel encodes all the trade-offs between properties (resolution, cross-term suppression, marginals, positivity).

---

**日本語.**  
Cohen（1966年）は、時間および周波数共変性を満たす**あらゆる**双線形時間周波数分布が以下の形で書けることを示しました：

$$\boxed{C_x(t, f) = \iint A_x(\eta, \tau)\, \Phi(\eta, \tau)\, e^{j2\pi(\eta t - f\tau)}\, d\eta\, d\tau}$$

- $A_x(\eta, \tau)$：$x$ の**対称アンビギュイティ関数**
- $\Phi(\eta, \tau)$：**カーネル関数**（アンビギュイティ領域でのフィルタ）

$\Phi$ の選び方によって **Cohenのクラス**の異なるメンバーが得られます。

## 4.2 Kernel Conditions for Desired Properties
## 4.2 所望の性質に対するカーネル条件

**English.**  

| Desired property | Kernel condition |
|-----------------|------------------|
| Real-valued $C_x$ | $\Phi(-\eta,-\tau) = \Phi^*(\eta,\tau)$ |
| Time marginal: $\int C_x\,df = |x(t)|^2$ | $\Phi(0, \tau) = 1\;\forall\tau$ |
| Frequency marginal: $\int C_x\,dt = |X(f)|^2$ | $\Phi(\eta, 0) = 1\;\forall\eta$ |
| Total energy: $\iint C_x = E$ | $\Phi(0, 0) = 1$ |
| Time-shift covariance | $|\Phi(\eta,\tau)|$ depends only on $\eta,\tau$ |
| Cross-term suppression | $\Phi(\eta,\tau) \to 0$ for large $|\eta|, |\tau|$ |

> **Fundamental trade-off:** The marginal conditions require $\Phi = 1$ on both axes, while cross-term suppression requires $\Phi \to 0$ away from the origin. No kernel can satisfy both perfectly — this is the kernel design problem.

---

**日本語.**  

| 所望の性質 | カーネル条件 |
|-----------|-------------|
| 実数値 | $\Phi(-\eta,-\tau) = \Phi^*(\eta,\tau)$ |
| 時間マージナル | $\Phi(0, \tau) = 1$ |
| 周波数マージナル | $\Phi(\eta, 0) = 1$ |
| 全エネルギー保存 | $\Phi(0, 0) = 1$ |
| クロスターム抑制 | $\Phi(\eta,\tau) \to 0$（大きな $|\eta|, |\tau|$ に対して） |

> **根本的なトレードオフ：** マージナル条件は両軸上で $\Phi = 1$ を要求しますが、クロスターム抑制は原点から離れると $\Phi \to 0$ を要求します。両方を完全に満たすカーネルは存在しません。

## 4.3 Important Members of Cohen's Class
## 4.3 Cohenのクラスの主要なメンバー

**English.**  

### Wigner–Ville Distribution
$$\Phi_{\text{WVD}}(\eta, \tau) = 1$$
Perfect marginals, maximum resolution, maximum cross-terms.

### Spectrogram (via window $h$)
$$\Phi_{\text{spec}}(\eta, \tau) = A_h(\eta, \tau)$$  
where $A_h$ is the ambiguity function of the analysis window.  
Non-negative, no cross-terms within $S_x$, but blurred resolution.

### Choi–Williams Distribution (CWD)
$$\Phi_{\text{CW}}(\eta, \tau) = \exp\!\left(-\frac{\eta^2 \tau^2}{\sigma}\right)$$
Satisfies marginals exactly ($\Phi(0,\tau)=\Phi(\eta,0)=1$), suppresses cross-terms for large $|\eta\tau|$. Parameter $\sigma > 0$ controls suppression strength.

### Born–Jordan Distribution
$$\Phi_{\text{BJ}}(\eta, \tau) = \frac{\sin(\pi\eta\tau)}{\pi\eta\tau} = \text{sinc}(\eta\tau)$$
Satisfies marginals, uniform averaging along hyperbolas.

### Rihaczek Distribution
$$\Phi_{\text{Ri}}(\eta, \tau) = e^{-j\pi\eta\tau}$$
$$R_x(t,f) = x(t)X^*(f)e^{j2\pi ft}$$
Complex-valued; exact marginals; strong cross-terms.

---

**日本語.**  

### Wigner–Ville 分布
$\Phi = 1$：完全なマージナル、最大分解能、最大クロスターム。

### スペクトログラム
$\Phi_{\text{spec}} = A_h(\eta, \tau)$（解析窓のアンビギュイティ関数）：非負、クロスタームなし、分解能は低い。

### Choi–Williams 分布（CWD）
$$\Phi_{\text{CW}}(\eta, \tau) = \exp\!\left(-\frac{\eta^2 \tau^2}{\sigma}\right)$$
マージナルを正確に満たし、大きな $|\eta\tau|$ でクロスタームを抑制。パラメータ $\sigma$ で抑制強度を制御。

### Born–Jordan 分布
$\Phi_{\text{BJ}}(\eta, \tau) = \text{sinc}(\eta\tau)$：マージナルを満たし、双曲線に沿った均一平均化。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert

plt.rcParams.update({'font.size': 11, 'figure.dpi': 120})
```

```python
# ─── Generic Cohen's Class computation via ambiguity domain ───────────────────
# アンビギュイティ領域を経由した汎用Cohenクラスの計算

def ambiguity_function(z, N_fft=None):
    """
    Compute the symmetric ambiguity function A_z(η, τ).
    z: analytic signal (complex array, length N)
    Returns A[η_index, τ_index], shape (N_fft, N)
    """
    N = len(z)
    if N_fft is None:
        N_fft = 2 * N

    # Instantaneous autocorrelation matrix: R[t, τ] = z(t+τ/2)*z*(t-τ/2)
    # Compute for integer lags on a grid
    R = np.zeros((N, N), dtype=complex)
    for lag in range(N):
        shift = lag
        t_valid = np.arange(shift, N - shift)
        if len(t_valid) == 0:
            break
        R[t_valid, lag] = z[t_valid + shift] * np.conj(z[t_valid - shift])
        if lag > 0:
            R[t_valid, N - lag] = np.conj(R[t_valid, lag])

    # A(η, τ): FT along time axis
    A = np.fft.fftshift(np.fft.fft(R, n=N_fft, axis=0), axes=0)
    return A


def cohen_class(z, kernel_fn, N_fft=None):
    """
    Compute a Cohen's class TFD given a kernel function.
    kernel_fn(eta_grid, tau_grid) → Phi(η, τ)
    Returns C[time, freq], shape (N, N_fft//2)
    """
    N = len(z)
    if N_fft is None:
        N_fft = 2 * N

    A = ambiguity_function(z, N_fft=N_fft)

    # Frequency (η) and lag (τ) grids
    eta = np.fft.fftshift(np.fft.fftfreq(N_fft))  # normalised Doppler
    tau = np.arange(N) / N                          # normalised lag
    ETA, TAU = np.meshgrid(eta, tau, indexing='ij')

    # Extend TAU to match full A shape (N_fft × N)
    Phi = kernel_fn(ETA, TAU)

    # Apply kernel and inverse 2D FT to get C(t, f)
    C_raw = np.fft.ifft2(np.fft.ifftshift(A * Phi, axes=0))
    C = np.real(C_raw[:N, :N_fft // 2])
    return C

print("Cohen's class functions defined / Cohenクラス関数を定義しました")
```

```python
# ─── Define kernel functions ───────────────────────────────────────────────────
# カーネル関数の定義

def kernel_wvd(eta, tau):
    """WVD kernel: Φ = 1"""
    return np.ones_like(eta)

def kernel_choi_williams(sigma=1.0):
    """Choi-Williams kernel: Φ = exp(-η²τ²/σ)"""
    def _kernel(eta, tau):
        return np.exp(-(eta * tau)**2 / sigma)
    return _kernel

def kernel_born_jordan(eta, tau):
    """Born-Jordan kernel: Φ = sinc(ητ)"""
    arg = np.pi * eta * tau
    return np.where(np.abs(arg) < 1e-10, 1.0, np.sin(arg) / arg)

print("Kernels defined:")
print("  Φ_WVD(0, 0) =", kernel_wvd(np.array([0.0]), np.array([0.0]))[0])
print("  Φ_CW(0, 0, σ=1) =", kernel_choi_williams(sigma=1.0)(np.array([0.0]), np.array([0.0]))[0])
print("  Φ_BJ(0, 0) =", kernel_born_jordan(np.array([0.0]), np.array([0.0]))[0])
```

```python
# ─── Visualise kernel shapes in the ambiguity domain ──────────────────────────
# アンビギュイティ領域でのカーネル形状の可視化

eta_1d = np.linspace(-2, 2, 400)
tau_1d = np.linspace(-2, 2, 400)
ETA, TAU = np.meshgrid(eta_1d, tau_1d)

kernels = {
    'WVD  (Φ = 1)': kernel_wvd(ETA, TAU),
    'Choi–Williams (σ=1)': kernel_choi_williams(1.0)(ETA, TAU),
    'Choi–Williams (σ=0.1)': kernel_choi_williams(0.1)(ETA, TAU),
    'Born–Jordan': kernel_born_jordan(ETA, TAU),
}

fig, axes = plt.subplots(1, 4, figsize=(18, 4))
fig.suptitle('Kernel functions Φ(η, τ) in the ambiguity domain\n'
             'アンビギュイティ領域でのカーネル関数 Φ(η, τ)', fontsize=12)

for ax, (name, K) in zip(axes, kernels.items()):
    im = ax.pcolormesh(eta_1d, tau_1d, K, cmap='RdBu_r',
                       vmin=-1, vmax=1, shading='gouraud')
    ax.set_title(name, fontsize=9)
    ax.set_xlabel('η (Doppler)')
    ax.set_ylabel('τ (lag)') if ax == axes[0] else None
    ax.set_aspect('equal')
    plt.colorbar(im, ax=ax, shrink=0.7)

plt.tight_layout()
plt.savefig('../figures/04_cohen_kernels.png', bbox_inches='tight')
plt.show()
```

```python
# ─── Compare TFDs on a two-component signal ───────────────────────────────────
# 2成分信号での各 TFD の比較

fs = 128
N  = 128
t  = np.arange(N) / fs

# Two Gaussian atoms
sigma_g = 0.06
a1 = np.exp(-((t - 0.3)**2) / (2*sigma_g**2)) * np.exp(1j * 2*np.pi * 20 * t)
a2 = np.exp(-((t - 0.7)**2) / (2*sigma_g**2)) * np.exp(1j * 2*np.pi * 45 * t)
z  = a1 + a2

N_fft = 256
freqs = np.linspace(0, fs / 2, N_fft // 2)

# Compute distributions
W_wvd = cohen_class(z, kernel_wvd, N_fft=N_fft)
W_cw1 = cohen_class(z, kernel_choi_williams(sigma=1.0), N_fft=N_fft)
W_cw2 = cohen_class(z, kernel_choi_williams(sigma=0.05), N_fft=N_fft)
W_bj  = cohen_class(z, kernel_born_jordan, N_fft=N_fft)

distributions = [
    ('WVD (Φ=1)', W_wvd),
    ('Choi–Williams (σ=1)', W_cw1),
    ('Choi–Williams (σ=0.05)', W_cw2),
    ('Born–Jordan', W_bj),
]

fig, axes = plt.subplots(1, 4, figsize=(18, 5), sharey=True)
fig.suptitle("Cohen's Class: TFD comparison on two Gaussian atoms\n"
             "Cohenのクラス：2つのガウスアトムに対する各TFDの比較", fontsize=12)

for ax, (name, W) in zip(axes, distributions):
    vmax = np.abs(W).max() * 0.85
    ax.pcolormesh(t, freqs, W.T, cmap='RdBu_r',
                  vmin=-vmax, vmax=vmax, shading='gouraud')
    ax.set_title(name, fontsize=9)
    ax.set_xlabel('Time (s)')
    ax.set_ylim(0, 70)

axes[0].set_ylabel('Frequency (Hz)')
plt.tight_layout()
plt.savefig('../figures/04_cohen_comparison.png', bbox_inches='tight')
plt.show()
```

## 4.4 Summary: The Cohen's Class Trade-off Table
## 4.4 まとめ：Cohenのクラスのトレードオフ表

| Distribution | Kernel $\Phi$ | Marginals | Cross-terms | Non-neg. | Resolution |
|---|---|---|---|---|---|
| WVD | 1 | ✅ exact | ❌ severe | ❌ | Optimal |
| Spectrogram | $A_h$ | ❌ blurred | ✅ none | ✅ | Low |
| Choi–Williams | $e^{-\eta^2\tau^2/\sigma}$ | ✅ exact | Partial | ❌ | Medium |
| Born–Jordan | sinc(ητ) | ✅ exact | Partial | ❌ | Medium |

**English.**  
Cohen's class provides a **complete parameterisation** of all covariant bilinear TFDs.  
The kernel $\Phi(\eta, \tau)$ acts as a **2-D filter in the ambiguity plane**: it passes the auto-terms (concentrated near the origin) while attenuating the cross-terms (which oscillate far from the origin).  
The optimal kernel depends on the application and the signal structure.

---

**日本語.**  
Cohenのクラスは、すべての共変双線形TFDの**完全なパラメータ化**を提供します。  
カーネル $\Phi(\eta, \tau)$ は**アンビギュイティ平面での2次元フィルタ**として機能します。原点付近に集中する自己項を通し、原点から遠く振動するクロスタームを減衰させます。  
最適なカーネルはアプリケーションと信号の構造に依存します。
