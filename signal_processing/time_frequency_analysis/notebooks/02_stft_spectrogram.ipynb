## Short-Time Fourier Transform (STFT) and the Spectrogram
## 短時間フーリエ変換（STFT）とスペクトログラム

---
> **Reference:** Cohen, *Time-Frequency Analysis*, Ch. 3; Gabor (1946).

## 2.1 Definition of the STFT
## 2.1 STFT の定義

**English.**  
The **Short-Time Fourier Transform** (STFT), introduced by Gabor (1946), localises the Fourier analysis in time by first multiplying $x(t)$ with a sliding analysis window $h(t)$ centred at time $\tau$:

$$\boxed{\mathcal{F}_x(\tau, f) = \int_{-\infty}^{\infty} x(t)\, h^*(t - \tau)\, e^{-j2\pi f t}\, dt}$$

- $h(t)$ — analysis window (real or complex, typically $\|h\|_2 = 1$)
- $\tau$ — time index (sliding position)
- $f$ — frequency
- $\mathcal{F}_x(\tau, f)$ — complex STFT coefficient

Alternatively, via the convolution perspective:

$$\mathcal{F}_x(\tau, f) = \left[ x(t)\, e^{-j2\pi f t} \ast h(-t) \right]_{t=\tau}$$

so the STFT is a **bandpass filter bank** applied to the down-mixed signal.

---

**日本語.**  
**短時間フーリエ変換（STFT）** はGabor（1946年）により導入されました。時刻 $\tau$ に中心を持つ解析窓 $h(t)$ で $x(t)$ を乗算することで、フーリエ解析を時間的に局在化します：

$$\boxed{\mathcal{F}_x(\tau, f) = \int_{-\infty}^{\infty} x(t)\, h^*(t - \tau)\, e^{-j2\pi f t}\, dt}$$

- $h(t)$ — 解析窓（実数または複素数、通常 $\|h\|_2 = 1$）
- $\tau$ — 時間インデックス（スライディング位置）
- $f$ — 周波数
- $\mathcal{F}_x(\tau, f)$ — 複素 STFT 係数

畳み込みの観点からは：STFT は down-mix された信号に適用する**帯域フィルタバンク**です。

## 2.2 The Spectrogram
## 2.2 スペクトログラム

**English.**  
The **spectrogram** is the squared magnitude of the STFT:

$$\boxed{S_x(\tau, f) = |\mathcal{F}_x(\tau, f)|^2}$$

It is **real and non-negative** — making it a physically appealing energy density.  
However, its marginals are **not** exact:

$$\int S_x(\tau, f)\, df = \int |x(t)|^2 |h(t-\tau)|^2\, dt \neq |x(\tau)|^2 \text{ (in general)}$$

The spectrogram can be expressed as the convolution of the Wigner distribution of $x$ and $h$:

$$S_x(\tau, f) = \iint W_x(t, \nu)\, W_h(\tau - t, f - \nu)\, dt\, d\nu$$

This shows that the spectrogram is a **smoothed Wigner distribution** — the window $h$ determines the smoothing kernel.

---

**日本語.**  
**スペクトログラム**は STFT の絶対値の二乗です：

$$\boxed{S_x(\tau, f) = |\mathcal{F}_x(\tau, f)|^2}$$

**実数値かつ非負**であるため、物理的に魅力的なエネルギー密度です。  
しかし、マージナルは**厳密ではありません**。

スペクトログラムは、$x$ と $h$ の Wigner 分布の畳み込みとして表現できます：

$$S_x(\tau, f) = \iint W_x(t, \nu)\, W_h(\tau - t, f - \nu)\, dt\, d\nu$$

これはスペクトログラムが**平滑化された Wigner 分布**であることを示しています。

## 2.3 Window Functions and the Resolution Trade-off
## 2.3 窓関数と分解能のトレードオフ

**English.**  
The choice of window $h(t)$ controls the **time-frequency resolution**:

| Window | Time resolution | Frequency resolution | Notes |
|--------|----------------|---------------------|-------|
| **Narrow** (short $\sigma$) | High | Low | Follows fast transients |
| **Wide** (large $\sigma$) | Low | High | Resolves close frequencies |

Common windows (for a signal of length $N$ samples):

$$h_{\text{Hann}}[n] = 0.5\left(1 - \cos\frac{2\pi n}{N-1}\right)$$

$$h_{\text{Gaussian}}(t) = \exp\!\left(-\frac{t^2}{2\sigma^2}\right) \quad \text{(optimal — achieves Heisenberg bound)}$$

$$h_{\text{Blackman}}[n] = 0.42 - 0.5\cos\frac{2\pi n}{N-1} + 0.08\cos\frac{4\pi n}{N-1}$$

---

**日本語.**  
窓 $h(t)$ の選択は**時間周波数分解能**を制御します：

| 窓 | 時間分解能 | 周波数分解能 | 備考 |
|-----|-----------|-------------|------|
| **狭い**（短い $\sigma$） | 高い | 低い | 速い過渡現象を追跡 |
| **広い**（大きい $\sigma$） | 低い | 高い | 近接周波数を分離 |

ガウス窓は不確定性原理の下限を達成するため最適です。

## 2.4 Inversion (Reconstruction)
## 2.4 逆変換（再構成）

**English.**  
The signal $x(t)$ can be reconstructed from its STFT provided $h(0) \neq 0$:

$$x(t) = \frac{1}{h^*(0)} \int_{-\infty}^{\infty} \mathcal{F}_x(\tau, f)\, e^{j2\pi f t}\, df$$

or equivalently via the **synthesis window** $\gamma(t)$:

$$x(t) = \iint \mathcal{F}_x(\tau, f)\, \gamma(t - \tau)\, e^{j2\pi f t}\, d\tau\, df$$

where $\langle \gamma, h \rangle = 1$ (bi-orthogonality condition).

---

**日本語.**  
$h(0) \neq 0$ の条件のもと、STFT から信号 $x(t)$ を再構成できます：

$$x(t) = \frac{1}{h^*(0)} \int_{-\infty}^{\infty} \mathcal{F}_x(\tau, f)\, e^{j2\pi f t}\, df$$

これは STFT が**可逆変換**であることを示しています。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft, get_window

plt.rcParams.update({'font.size': 11, 'figure.dpi': 120})
rng = np.random.default_rng(42)
```

```python
# ─── Generate a non-stationary test signal ────────────────────────────────────
# 非定常テスト信号の生成
# Signal: linear chirp + pure tone + noise burst
# 信号：線形チャープ ＋ 純音 ＋ ノイズバースト

fs = 1000          # Sampling rate [Hz] / サンプリング周波数
T  = 2.0           # Duration [s] / 持続時間
t  = np.arange(0, T, 1/fs)

# 1) Linear chirp: 50 Hz → 300 Hz over [0, 1s]
f_start, f_end = 50.0, 300.0
chirp = np.where(t < 1.0,
                 np.cos(2*np.pi * (f_start*t + (f_end - f_start)/(2*1.0) * t**2)),
                 0.0)

# 2) Pure tone at 150 Hz active during [0.8s, 1.8s]
tone = np.where((t >= 0.8) & (t <= 1.8),
                0.7 * np.cos(2*np.pi * 150 * t),
                0.0)

# 3) White noise burst during [1.5s, 1.7s]
noise = np.where((t >= 1.5) & (t <= 1.7),
                  0.3 * rng.standard_normal(len(t)),
                  0.0)

x = chirp + tone + noise

plt.figure(figsize=(12, 2.5))
plt.plot(t, x, lw=0.7)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Test signal: chirp + tone + noise burst\n'
          'テスト信号：チャープ + 純音 + ノイズバースト')
plt.tight_layout()
plt.show()
```

```python
# ─── Compare window shapes and their spectra ──────────────────────────────────
# 窓関数の形状とスペクトルの比較

M = 256   # window length
windows = {
    'Rectangular (矩形)':  np.ones(M),
    'Hann (ハン)':          get_window('hann', M),
    'Blackman (ブラックマン)': get_window('blackman', M),
    'Gaussian σ=0.25 (ガウス)': get_window(('gaussian', M * 0.25 / 2), M),
}

fig, axes = plt.subplots(2, 4, figsize=(16, 6))
fig.suptitle('Window functions (time domain) and their spectra\n'
             '窓関数（時間領域）とそのスペクトル', fontsize=12)

for col, (name, w) in enumerate(windows.items()):
    n = np.arange(M)
    W = np.abs(np.fft.fftshift(np.fft.fft(w, n=4096)))
    freq = np.linspace(-0.5, 0.5, 4096)

    axes[0, col].plot(n, w, lw=1.5)
    axes[0, col].set_title(name, fontsize=9)
    axes[0, col].set_xlabel('Sample')
    axes[0, col].set_ylabel('Amplitude') if col == 0 else None

    axes[1, col].plot(freq, 20 * np.log10(W / W.max() + 1e-12), lw=1.2)
    axes[1, col].set_ylim(-80, 5)
    axes[1, col].set_xlim(-0.15, 0.15)
    axes[1, col].set_xlabel('Normalised frequency')
    axes[1, col].set_ylabel('Magnitude (dB)') if col == 0 else None
    axes[1, col].axhline(-3, color='r', ls='--', lw=0.8, label='-3 dB')

plt.tight_layout()
plt.show()
```

```python
# ─── STFT spectrograms with different window lengths ──────────────────────────
# 異なる窓長での STFT スペクトログラム

nperseg_values = [32, 128, 256, 512]

fig, axes = plt.subplots(1, 4, figsize=(18, 4), sharey=True)
fig.suptitle('Spectrogram resolution trade-off (Hann window)\n'
             'スペクトログラムの分解能トレードオフ（ハン窓）', fontsize=12)

for ax, nperseg in zip(axes, nperseg_values):
    f_ax, t_ax, Zxx = stft(x, fs=fs, window='hann', nperseg=nperseg,
                            noverlap=nperseg * 3 // 4)
    S = 20 * np.log10(np.abs(Zxx) + 1e-10)
    S -= S.max()
    im = ax.pcolormesh(t_ax, f_ax, S, vmin=-50, vmax=0,
                       cmap='inferno', shading='gouraud')
    ax.set_ylim(0, 400)
    ax.set_xlabel('Time (s)')
    ax.set_title(f'N_win = {nperseg}\nΔt ≈ {nperseg/2/fs*1000:.0f} ms, '
                 f'Δf ≈ {fs/nperseg:.0f} Hz')

axes[0].set_ylabel('Frequency (Hz)')
plt.colorbar(im, ax=axes[-1], label='Magnitude (dB)')
plt.tight_layout()
plt.savefig('../figures/02_stft_resolution.png', bbox_inches='tight')
plt.show()
```

```python
# ─── Compare global Fourier spectrum vs STFT ──────────────────────────────────
# グローバルフーリエスペクトルと STFT の比較

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Global power spectrum
X_global = np.fft.rfft(x)
freqs = np.fft.rfftfreq(len(x), 1/fs)
ax1.semilogy(freqs, np.abs(X_global)**2 / len(x))
ax1.set_xlabel('Frequency (Hz)')
ax1.set_ylabel('Power')
ax1.set_xlim(0, 400)
ax1.set_title('Global Power Spectrum (Fourier)\nグローバルパワースペクトル（フーリエ）')
ax1.axvline(50, color='g', ls='--', alpha=0.7, label='50 Hz chirp start')
ax1.axvline(150, color='r', ls='--', alpha=0.7, label='150 Hz tone')
ax1.axvline(300, color='b', ls='--', alpha=0.7, label='300 Hz chirp end')
ax1.legend(fontsize=8)

# Optimal STFT
f_ax, t_ax, Zxx = stft(x, fs=fs, window='hann', nperseg=256, noverlap=240)
S = 20 * np.log10(np.abs(Zxx) + 1e-10)
S -= S.max()
im = ax2.pcolormesh(t_ax, f_ax, S, vmin=-40, vmax=0, cmap='magma', shading='gouraud')
ax2.set_ylim(0, 400)
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Frequency (Hz)')
ax2.set_title('STFT Spectrogram (N=256, Hann)\nSTFT スペクトログラム')
plt.colorbar(im, ax=ax2, label='dB')

plt.tight_layout()
plt.show()
```

## 2.5 Summary
## 2.5 まとめ

| Property | STFT / Spectrogram |
|----------|--------------------|
| Definition | $\mathcal{F}_x(\tau,f) = \int x(t)h^*(t-\tau)e^{-j2\pi ft}dt$ |
| TFD | $S_x(\tau,f) = |\mathcal{F}_x(\tau,f)|^2$ |
| Non-negative? | ✅ Yes |
| Exact marginals? | ❌ No (smoothed by $\|h\|^2$) |
| Cross-terms? | ✅ None (linear transform) |
| Resolution trade-off | Controlled by window length $M$ |

**English.**  
The spectrogram is the most widely used TFD in practice due to its simplicity and guaranteed non-negativity.  
Its chief limitation is the **Heisenberg resolution box**: the product $\Delta_t \cdot \Delta_f$ cannot be made smaller than $1/(4\pi)$, regardless of the window choice.  
The Wigner distribution (Chapter 3) achieves far better resolution at the cost of introducing cross-terms.

---

**日本語.**  
スペクトログラムはそのシンプルさと非負性の保証から、実用上最も広く使われる TFD です。  
主な制限は**ハイゼンベルク分解能ボックス**です：窓の選択に関わらず、$\Delta_t \cdot \Delta_f$ を $1/(4\pi)$ より小さくすることはできません。  
第3章の Wigner 分布は、クロスターム（交差項）の代償として格段に優れた分解能を達成します。
