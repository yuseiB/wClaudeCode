## Instantaneous Frequency and the Analytic Signal
## 瞬時周波数とアナリティック信号

---
> **Reference:** Cohen, *Time-Frequency Analysis*, Ch. 2; Gabor (1946); Ville (1948).

## 5.1 The Analytic Signal
## 5.1 アナリティック信号

**English.**  
For a real signal $x(t)$, define the **analytic signal** $z(t)$ as:

$$\boxed{z(t) = x(t) + j\hat{x}(t) = A(t)\, e^{j\phi(t)}}$$

where $\hat{x}(t)$ is the **Hilbert transform** of $x(t)$:

$$\hat{x}(t) = \mathcal{H}\{x\}(t) = \frac{1}{\pi}\, \text{P.V.}\int_{-\infty}^{\infty} \frac{x(\tau)}{t - \tau}\, d\tau$$

In frequency domain, the Hilbert transform multiplies positive frequencies by $-j$ and negative by $+j$:

$$\hat{X}(f) = -j\, \text{sgn}(f)\, X(f)$$

Therefore the analytic signal has a **one-sided spectrum**:

$$Z(f) = \begin{cases} 2X(f) & f > 0 \\ X(0) & f = 0 \\ 0 & f < 0 \end{cases}$$

The analytic signal provides unambiguous definitions of:
- **Instantaneous amplitude (envelope):** $A(t) = |z(t)| = \sqrt{x(t)^2 + \hat{x}(t)^2}$
- **Instantaneous phase:** $\phi(t) = \arg z(t) = \arctan\!\left(\frac{\hat{x}(t)}{x(t)}\right)$
- **Instantaneous frequency:** $f_i(t) = \frac{1}{2\pi}\frac{d\phi}{dt}$

---

**日本語.**  
実信号 $x(t)$ に対して、**アナリティック信号** $z(t)$ を以下のように定義します：

$$\boxed{z(t) = x(t) + j\hat{x}(t) = A(t)\, e^{j\phi(t)}}$$

ここで $\hat{x}(t)$ は $x(t)$ の**Hilbert変換**です。

周波数領域では、アナリティック信号は**片側スペクトル**を持ちます。

アナリティック信号は以下の明確な定義を与えます：
- **瞬時振幅（エンベロープ）:** $A(t) = |z(t)|$
- **瞬時位相:** $\phi(t) = \arg z(t)$
- **瞬時周波数:** $f_i(t) = \frac{1}{2\pi}\frac{d\phi}{dt}$

## 5.2 Instantaneous Frequency
## 5.2 瞬時周波数

**English.**  
The **instantaneous frequency (IF)** is defined as the time derivative of the phase of the analytic signal:

$$\boxed{f_i(t) = \frac{1}{2\pi}\frac{d\phi(t)}{dt}}$$

**Key examples:**

1. **Pure tone:** $x(t) = A\cos(2\pi f_0 t)$  
   $z(t) = Ae^{j2\pi f_0 t}$, $\phi(t) = 2\pi f_0 t$, $f_i(t) = f_0$ ✓

2. **Linear chirp:** $x(t) = \cos\!\left(2\pi\left(f_0 t + \frac{k}{2}t^2\right)\right)$  
   $\phi(t) = 2\pi\left(f_0 t + \frac{k}{2}t^2\right)$, $f_i(t) = f_0 + kt$ ✓

3. **FM signal:** $x(t) = \cos\!\left(2\pi f_c t + \beta\sin(2\pi f_m t)\right)$  
   $f_i(t) = f_c + \beta f_m \cos(2\pi f_m t)$ ✓

**Connection to WVD:** The conditional mean frequency of the WVD equals the instantaneous frequency:

$$\bar{f}(t) \equiv \frac{\int f\, W_z(t, f)\, df}{\int W_z(t, f)\, df} = f_i(t)$$

---

**日本語.**  
**瞬時周波数（IF）** はアナリティック信号の位相の時間微分として定義されます：

$$\boxed{f_i(t) = \frac{1}{2\pi}\frac{d\phi(t)}{dt}}$$

**主要な例：**
1. **純音：** $f_i = f_0$（定数）
2. **線形チャープ：** $f_i(t) = f_0 + kt$（線形増加）
3. **FM信号：** $f_i(t) = f_c + \beta f_m \cos(2\pi f_m t)$

**WVD との関係：** WVD の条件付き平均周波数は瞬時周波数に等しい。

## 5.3 Conditions for a Meaningful Instantaneous Frequency
## 5.3 意味のある瞬時周波数の条件

**English.**  
The IF is only physically meaningful for **monocomponent signals** (signals with a single "ridge" in the TF plane).  
For a multicomponent signal $x = x_1 + x_2$, the IF of $x$ is a weighted average of the IFs of the components:

$$f_i(t) = \frac{A_1^2(t)\, f_{i,1}(t) + A_2^2(t)\, f_{i,2}(t)}{A_1^2(t) + A_2^2(t)}$$

which can take values **outside** the range $[f_{i,1}(t),\, f_{i,2}(t)]$ and may even be negative.

**Bedrosian's theorem:** For the analytic signal to correctly factor as $z(t) = A(t)e^{j\phi(t)}$, the spectra of the envelope $A(t)$ and the carrier $e^{j\phi(t)}$ must not overlap.

---

**日本語.**  
瞬時周波数が物理的に意味を持つのは、TF平面上に単一の「リッジ」を持つ**単成分信号**のみです。  
多成分信号では、各成分の瞬時周波数の**重み付き平均**となり、物理的意味を失います。  

**Bedrosianの定理：** アナリティック信号が $z(t) = A(t)e^{j\phi(t)}$ と正しく分解されるためには、エンベロープ $A(t)$ と搬送波 $e^{j\phi(t)}$ のスペクトルが重ならないことが必要です。

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import hilbert

plt.rcParams.update({'font.size': 11, 'figure.dpi': 120})
```

```python
# ─── Analytic signal and envelope extraction ──────────────────────────────────
# アナリティック信号とエンベロープ抽出

fs = 1000
t  = np.linspace(0, 1.0, fs, endpoint=False)

# AM signal: carrier 80 Hz, modulated by 5 Hz sinusoid
# AM信号：80 Hz 搬送波、5 Hz 正弦波で変調
A_true = 1 + 0.6 * np.cos(2 * np.pi * 5 * t)   # true envelope
x_am   = A_true * np.cos(2 * np.pi * 80 * t)

# Analytic signal via Hilbert transform
z = hilbert(x_am)
A_hat = np.abs(z)           # estimated envelope
phi   = np.unwrap(np.angle(z))  # instantaneous phase
fi    = np.diff(phi) / (2 * np.pi) * fs  # instantaneous frequency [Hz]

fig, axes = plt.subplots(3, 1, figsize=(12, 9), sharex=True)
fig.suptitle('Analytic signal analysis of an AM signal\nAM信号のアナリティック信号解析', fontsize=12)

axes[0].plot(t, x_am, lw=0.7, label='AM signal $x(t)$')
axes[0].plot(t, A_true, 'r-', lw=2, label='True envelope $A(t)$')
axes[0].plot(t, A_hat, 'g--', lw=1.5, label='Hilbert envelope $|z(t)|$')
axes[0].plot(t, -A_true, 'r-', lw=2, alpha=0.5)
axes[0].set_ylabel('Amplitude')
axes[0].legend(loc='upper right', fontsize=9)

axes[1].plot(t, phi / (2*np.pi), lw=1.2, color='purple')
axes[1].set_ylabel('Phase / cycles')
axes[1].set_title('Instantaneous phase φ(t) / 瞬時位相')

axes[2].plot(t[1:], fi, lw=1.2, color='darkorange')
axes[2].axhline(80, color='k', ls='--', lw=1, label='Carrier 80 Hz')
axes[2].set_ylabel('Frequency (Hz)')
axes[2].set_xlabel('Time (s)')
axes[2].set_title('Instantaneous frequency $f_i(t)$ / 瞬時周波数')
axes[2].set_ylim(60, 100)
axes[2].legend()

plt.tight_layout()
plt.savefig('../figures/05_analytic_am.png', bbox_inches='tight')
plt.show()
```

```python
# ─── Instantaneous frequency of various signal types ──────────────────────────
# 様々な信号タイプの瞬時周波数

fs = 2000
t  = np.linspace(0, 1.0, fs, endpoint=False)

signals = {}

# 1. Linear chirp
f0, f1 = 20.0, 200.0
signals['Linear Chirp\n線形チャープ'] = (
    np.cos(2*np.pi*(f0*t + (f1-f0)/(2*1.0)*t**2)),
    f0 + (f1-f0)*t   # true IF
)

# 2. Quadratic chirp
signals['Quadratic Chirp\n二次チャープ'] = (
    np.cos(2*np.pi*(50*t + 30*t**2 + 50*t**3)),
    50 + 60*t + 150*t**2  # true IF
)

# 3. FM signal
fc, beta, fm = 100.0, 50.0, 5.0
signals['FM Signal\nFM信号'] = (
    np.cos(2*np.pi*fc*t + beta*np.sin(2*np.pi*fm*t)),
    fc + beta*fm*np.cos(2*np.pi*fm*t)  # true IF
)

fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Instantaneous frequency estimation via Hilbert transform\n'
             'Hilbert変換による瞬時周波数の推定', fontsize=12)

for ax, (name, (x, fi_true)) in zip(axes, signals.items()):
    z = hilbert(x)
    phi_est = np.unwrap(np.angle(z))
    fi_est  = np.diff(phi_est) / (2*np.pi) * fs

    ax.plot(t[1:], fi_true[1:], 'k-', lw=2.5, label='True IF')
    ax.plot(t[1:], fi_est,  'r--', lw=1.5, alpha=0.85, label='Estimated IF')
    ax.set_xlabel('Time (s)')
    ax.set_ylabel('Frequency (Hz)')
    ax.set_title(name)
    ax.set_ylim(0, fs // 4)
    ax.legend(fontsize=9)

plt.tight_layout()
plt.savefig('../figures/05_instantaneous_freq.png', bbox_inches='tight')
plt.show()
```

```python
# ─── Failure of IF for multicomponent signals ──────────────────────────────────
# 多成分信号での瞬時周波数の失敗

fs = 2000
t  = np.linspace(0, 1.0, fs, endpoint=False)

x1 = np.cos(2*np.pi * 50  * t)
x2 = np.cos(2*np.pi * 150 * t)
x  = x1 + x2   # two-tone signal

z    = hilbert(x)
phi  = np.unwrap(np.angle(z))
fi   = np.diff(phi) / (2*np.pi) * fs

fig, axes = plt.subplots(2, 1, figsize=(12, 7), sharex=True)
fig.suptitle('IF failure for multicomponent signal (two tones)\n'
             '多成分信号（2つの正弦波）での瞬時周波数の失敗', fontsize=12)

axes[0].plot(t, x, lw=0.8, label='$x(t) = \cos(2\pi\cdot50t) + \cos(2\pi\cdot150t)$')
axes[0].plot(t, np.abs(z), 'r-', lw=1.5, label='Envelope $|z(t)|$')
axes[0].set_ylabel('Amplitude')
axes[0].legend(fontsize=9)

axes[1].plot(t[1:], fi, lw=1.0, color='darkorange')
axes[1].axhline(50,  color='b', ls='--', lw=1.5, label='50 Hz', alpha=0.7)
axes[1].axhline(150, color='g', ls='--', lw=1.5, label='150 Hz', alpha=0.7)
axes[1].axhline(100, color='k', ls=':', lw=1,
                label='Mean = 100 Hz (meaningless!)', alpha=0.5)
axes[1].set_ylim(-100, 300)
axes[1].set_xlabel('Time (s)')
axes[1].set_ylabel('IF (Hz)')
axes[1].set_title('Estimated IF — oscillates between components\n'
                  '推定瞬時周波数：成分間を振動（物理的意味なし）')
axes[1].legend(fontsize=9)

plt.tight_layout()
plt.savefig('../figures/05_multicomp_if_failure.png', bbox_inches='tight')
plt.show()

print("\nConclusion / 結論:")
print("IF is only meaningful for monocomponent signals!")
print("瞬時周波数が意味を持つのは単成分信号のみです！")
```

## 5.4 Summary
## 5.4 まとめ

| Concept | Formula |
|---------|--------|
| Hilbert transform | $\hat{x}(t) = \frac{1}{\pi}\text{P.V.}\int\frac{x(\tau)}{t-\tau}d\tau$ |
| Analytic signal | $z(t) = x(t) + j\hat{x}(t) = A(t)e^{j\phi(t)}$ |
| Instantaneous frequency | $f_i(t) = \frac{1}{2\pi}\frac{d\phi}{dt}$ |
| Validity condition | Monocomponent (Bedrosian's theorem) |
| Connection to WVD | $\bar{f}(t) = \int f W_z\,df \big/ \int W_z\,df = f_i(t)$ |

**English.**  
The analytic signal is the foundation of modern time-frequency analysis.  
By restricting to positive frequencies, it eliminates aliasing in TFDs and makes the IF definition unique.  
For multicomponent signals, decomposition methods (EMD, SST, ridge extraction) are needed before applying IF estimation.

---

**日本語.**  
アナリティック信号は現代の時間周波数解析の基礎です。  
正周波数のみに制限することで、TFD のエイリアシングを排除し、瞬時周波数の定義を一意にします。  
多成分信号では、瞬時周波数推定を適用する前に分解法（EMD、SST、リッジ抽出）が必要です。
