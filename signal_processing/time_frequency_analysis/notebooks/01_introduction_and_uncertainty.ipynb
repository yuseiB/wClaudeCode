## 1.1 Motivation — Why Time-Frequency Analysis?
## 1.1 動機 — なぜ時間周波数解析が必要か？

**English.**  
Classical Fourier analysis decomposes a signal $x(t)$ into frequency components:

$$X(f) = \int_{-\infty}^{\infty} x(t)\, e^{-j2\pi f t}\, dt$$

This tells us *which* frequencies are present globally, but gives **no information about when** they occur.  
Real-world signals — speech, music, radar echoes, seismic waves, biomedical signals — are **non-stationary**: their frequency content evolves with time.

| Signal | Reason for non-stationarity |
|--------|-----------------------------|
| Speech | Formants shift as phonemes change |
| ECG / EEG | Rhythms modulated by physiological state |
| Bat sonar | Frequency-modulated pulse |
| Gravitational wave | Chirping inspiral |

The central problem of time-frequency analysis is to define a function $\rho(t, f)$ — called a **time-frequency distribution (TFD)** — that answers:

> *"How much energy is concentrated around frequency $f$ at time $t$?"*

---

**日本語.**  
古典的なフーリエ解析はシグナル $x(t)$ を周波数成分に分解します：

$$X(f) = \int_{-\infty}^{\infty} x(t)\, e^{-j2\pi f t}\, dt$$

これはどの周波数が*全体として*存在するかは教えてくれますが、**それがいつ現れるかの情報はありません**。  
音声・音楽・レーダー・地震波・生体信号などの実世界の信号は**非定常**であり、その周波数内容は時間とともに変化します。

時間周波数解析の中心的問題は、**時間周波数分布（TFD）** と呼ばれる関数 $\rho(t, f)$ を定義することです：

> *「時刻 $t$ において、どれだけのエネルギーが周波数 $f$ 付近に集中しているか？」*

## 1.2 Desirable Properties of a TFD
## 1.2 TFD に求められる性質

**English.**  
An ideal TFD $\rho(t, f)$ should satisfy:

1. **Real-valued** and non-negative: $\rho(t,f) \geq 0$  
2. **Time marginal** — integrating over $f$ gives the instantaneous power:  
   $$\int_{-\infty}^{\infty} \rho(t, f)\, df = |x(t)|^2$$
3. **Frequency marginal** — integrating over $t$ gives the energy spectrum:  
   $$\int_{-\infty}^{\infty} \rho(t, f)\, dt = |X(f)|^2$$
4. **Total energy**:  
   $$\iint \rho(t, f)\, dt\, df = \int |x(t)|^2\, dt = E$$
5. **Time covariance** — shifting $x(t)$ in time shifts $\rho$ accordingly.  
6. **Frequency covariance** — modulating $x(t)$ shifts $\rho$ in frequency.

> **Wigner's theorem (1932):** No distribution can satisfy all of the above simultaneously, especially non-negativity alongside correct marginals. This is a consequence of the quantum-mechanical uncertainty principle.

---

**日本語.**  
理想的な TFD $\rho(t, f)$ は以下を満たすべきです：

1. **実数値**かつ非負：$\rho(t,f) \geq 0$  
2. **時間マージナル**（周波数方向に積分すると瞬時パワーになる）：  
   $$\int_{-\infty}^{\infty} \rho(t, f)\, df = |x(t)|^2$$
3. **周波数マージナル**（時間方向に積分するとエネルギースペクトルになる）：  
   $$\int_{-\infty}^{\infty} \rho(t, f)\, dt = |X(f)|^2$$
4. **全エネルギー**：  
   $$\iint \rho(t, f)\, dt\, df = E$$
5. **時間共変性**：$x(t)$ を時間シフトすると $\rho$ も対応してシフトする。  
6. **周波数共変性**：$x(t)$ を変調すると $\rho$ は周波数方向にシフトする。

> **Wignerの定理（1932年）：** 非負性と正しいマージナルを同時に満たすような分布は存在しません。これは量子力学の不確定性原理の帰結です。

## 1.3 The Heisenberg–Gabor Uncertainty Principle
## 1.3 ハイゼンベルク–ガボール不確定性原理

**English.**  
Define the **duration** $\Delta_t$ and **bandwidth** $\Delta_f$ of a signal by:

$$\Delta_t^2 = \frac{\int (t - \bar{t})^2 |x(t)|^2\, dt}{E}, \qquad \bar{t} = \frac{\int t|x(t)|^2\, dt}{E}$$

$$\Delta_f^2 = \frac{\int (f - \bar{f})^2 |X(f)|^2\, df}{E}, \qquad \bar{f} = \frac{\int f|X(f)|^2\, df}{E}$$

The **uncertainty principle** states:

$$\boxed{\Delta_t \cdot \Delta_f \geq \frac{1}{4\pi}}$$

The bound is achieved by the **Gaussian** (Gabor) signal:

$$g(t) = \left(\frac{1}{2\pi\sigma^2}\right)^{1/4} \exp\!\left(-\frac{t^2}{4\sigma^2} + j2\pi f_0 t\right)$$

This has $\Delta_t = \sigma$ and $\Delta_f = \frac{1}{4\pi\sigma}$, so $\Delta_t \Delta_f = \frac{1}{4\pi}$.

**Physical interpretation:** A signal cannot be simultaneously arbitrarily concentrated in both time and frequency — a short pulse necessarily has a wide spectrum, and a pure tone necessarily extends indefinitely in time.

---

**日本語.**  
信号の**継続時間** $\Delta_t$ と**帯域幅** $\Delta_f$ を以下で定義します：

$$\Delta_t^2 = \frac{\int (t - \bar{t})^2 |x(t)|^2\, dt}{E}, \qquad \bar{t} = \frac{\int t|x(t)|^2\, dt}{E}$$

$$\Delta_f^2 = \frac{\int (f - \bar{f})^2 |X(f)|^2\, df}{E}, \qquad \bar{f} = \frac{\int f|X(f)|^2\, df}{E}$$

**不確定性原理：**

$$\boxed{\Delta_t \cdot \Delta_f \geq \frac{1}{4\pi}}$$

この下限は**ガウス信号**（Gabor信号）によって達成されます：

$$g(t) = \left(\frac{1}{2\pi\sigma^2}\right)^{1/4} \exp\!\left(-\frac{t^2}{4\sigma^2} + j2\pi f_0 t\right)$$

**物理的解釈：** 信号は時間と周波数の両方で同時に任意に集中させることはできません。短いパルスは広いスペクトルを持ち、純音は時間的に無限に広がります。

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

plt.rcParams.update({'font.size': 12, 'figure.dpi': 120})
```

```python
# ─── Uncertainty Principle: Gaussian achieves the bound ───────────────────────
# 不確定性原理：ガウス信号は下限を達成する

def gaussian_signal(t, sigma, f0=0.0):
    """Gabor / Gaussian signal centred at t=0, f=f0."""
    norm = (1 / (2 * np.pi * sigma**2)) ** 0.25
    return norm * np.exp(-t**2 / (4 * sigma**2)) * np.exp(1j * 2 * np.pi * f0 * t)

def time_bandwidth_product(x, t):
    """Compute Δt, Δf, and their product for signal x sampled on grid t."""
    dt = t[1] - t[0]
    E = np.trapz(np.abs(x)**2, t)

    t_mean = np.trapz(t * np.abs(x)**2, t) / E
    dt_rms = np.sqrt(np.trapz((t - t_mean)**2 * np.abs(x)**2, t) / E)

    N = len(x)
    fs = 1.0 / dt
    f = np.fft.fftshift(np.fft.fftfreq(N, d=dt))
    X = np.fft.fftshift(np.fft.fft(x)) * dt

    f_mean = np.trapz(f * np.abs(X)**2, f) / E
    df_rms = np.sqrt(np.trapz((f - f_mean)**2 * np.abs(X)**2, f) / E)

    return dt_rms, df_rms, dt_rms * df_rms

# Grid
t = np.linspace(-4, 4, 4096)
sigmas = [0.3, 0.6, 1.0, 1.5]

print(f"{'σ':>6}  {'Δt':>8}  {'Δf':>8}  {'Δt·Δf':>10}  {'bound 1/4π':>12}")
print("-" * 56)
for σ in sigmas:
    x = gaussian_signal(t, sigma=σ)
    dt_rms, df_rms, tbp = time_bandwidth_product(x, t)
    print(f"{σ:>6.2f}  {dt_rms:>8.4f}  {df_rms:>8.4f}  {tbp:>10.6f}  {1/(4*np.pi):>12.6f}")
```

```python
# ─── Visualise the uncertainty trade-off ──────────────────────────────────────
# 不確定性のトレードオフを可視化する

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Uncertainty Principle: Gaussian signals with different σ\n"
             "不確定性原理：異なる σ のガウス信号", fontsize=13)

for ax, σ in zip(axes.flat, sigmas):
    x = gaussian_signal(t, sigma=σ)
    N = len(x)
    dt = t[1] - t[0]
    f = np.fft.fftshift(np.fft.fftfreq(N, d=dt))
    X = np.fft.fftshift(np.fft.fft(x)) * dt

    dt_rms, df_rms, tbp = time_bandwidth_product(x, t)

    color_t, color_f = '#2196F3', '#FF5722'
    ax2 = ax.twinx()
    ax.plot(t, np.abs(x)**2, color=color_t, lw=1.8, label=r'$|x(t)|^2$')
    ax2.plot(f, np.abs(X)**2, color=color_f, lw=1.8, alpha=0.85, label=r'$|X(f)|^2$')

    ax.set_xlim(-4, 4)
    ax2.set_xlim(-4, 4)   # same axis for overlay clarity
    ax.set_xlabel('Time  t  /  Frequency  f')
    ax.set_ylabel(r'$|x(t)|^2$', color=color_t)
    ax2.set_ylabel(r'$|X(f)|^2$', color=color_f)
    ax.set_title(f'σ = {σ:.1f}   |   Δt·Δf = {tbp:.4f}  (≥ {1/(4*np.pi):.4f})')

    lines1, labels1 = ax.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax.legend(lines1 + lines2, labels1 + labels2, fontsize=9)

plt.tight_layout()
plt.savefig('../figures/01_uncertainty_gaussian.png', bbox_inches='tight')
plt.show()
```
